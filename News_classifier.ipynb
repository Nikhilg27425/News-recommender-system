```python
import pandas as pd

col_names = [
    "newsId", "category", "subCategory", "title",
    "abstract", "url", "titleEntities", "abstractEntities"
]

df = pd.read_csv("/content/mind_news/news.tsv/news.tsv", sep="\t", names=col_names, header=None)
print(df.shape)
df.head()
```

```python
print("\nColumns:", df.columns.tolist())
print("\nUnique Categories:", df['category'].unique())
print("\nNumber of categories:", df['category'].nunique())
print("\nMissing values:\n", df.isna().sum())
```

```python
df.info()
print("\nUnique Categories:\n", df['category'].unique())
print("\nUnique Subcategories:\n", df['subCategory'].unique())
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder

df['text'] = (df['title'].fillna('') + ' ' + df['abstract'].fillna('')).str.strip()

le = LabelEncoder()
df['label'] = le.fit_transform(df['category'])

print("Encoded Categories:", dict(zip(le.classes_, le.transform(le.classes_))))
```

```python
cat_counts = df['category'].value_counts()
rare_cats = cat_counts[cat_counts < 2].index

print("Dropping these rare categories:", list(rare_cats))

df = df[~df['category'].isin(rare_cats)]
print("New dataset size:", df.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

le = LabelEncoder()
df['label'] = le.fit_transform(df['category'])

X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['label'], test_size=0.2, random_state=42, stratify=df['label']
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    max_features=20000,
    ngram_range=(1,2),
    stop_words='english'
)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf  = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=200, n_jobs=-1)
model.fit(X_train_tfidf, y_train)
```

```python
from sklearn.metrics import classification_report
import numpy as np

probs = model.predict_proba(X_test_tfidf)

category_names = le.classes_

for i in range(5):
    high_conf = [(category_names[j], round(probs[i][j],3))
                 for j in range(len(category_names)) if probs[i][j] > 0.5]
    print(f"\nText {i+1}: {high_conf if high_conf else 'None > 0.5'}")

pred_labels = np.argmax(probs, axis=1)
print("\nClassification Report:\n")
print(classification_report(
    y_test,
    pred_labels,
    labels=np.arange(len(le.classes_)),
    target_names=le.classes_,
    zero_division=0
))
```

Multi classes classifier pipeline


```python
import pandas as pd
import numpy as np
import pickle
import joblib
from datetime import datetime

# ============================================
# 1. SAVE TRAINED MODEL
# ============================================

def save_model(model, mlb, vectorizer, filepath='news_classifier_model.pkl'):
    """
    Save trained model, label encoder, and vectorizer

    Parameters:
    - model: Trained MultiOutputClassifier
    - mlb: MultiLabelBinarizer
    - vectorizer: TfidfVectorizer
    - filepath: Path to save model
    """

    model_package = {
        'model': model,
        'mlb': mlb,
        'vectorizer': vectorizer,
        'saved_at': datetime.now().isoformat(),
        'classes': list(mlb.classes_)
    }

    try:
        with open(filepath, 'wb') as f:
            pickle.dump(model_package, f)
        print(f" Model saved successfully to {filepath}")
        print(f"  Classes: {model_package['classes']}")
    except Exception as e:
        print(f" Error saving model: {e}")

# ============================================
# 2. LOAD TRAINED MODEL
# ============================================

def load_model(filepath='news_classifier_model.pkl'):
    """
    Load trained model, label encoder, and vectorizer

    Parameters:
    - filepath: Path to saved model

    Returns:
    - model, mlb, vectorizer
    """

    try:
        with open(filepath, 'rb') as f:
            model_package = pickle.load(f)

        model = model_package['model']
        mlb = model_package['mlb']
        vectorizer = model_package['vectorizer']

        print(f" Model loaded successfully from {filepath}")
        print(f"  Saved at: {model_package['saved_at']}")
        print(f"  Classes: {model_package['classes']}")

        return model, mlb, vectorizer

    except Exception as e:
        print(f" Error loading model: {e}")
        return None, None, None

# ============================================
# 3. PREPROCESS NEW ARTICLES FOR PREDICTION
# ============================================

def preprocess_for_prediction(df):
    """
    Preprocess new articles (same as training preprocessing)

    Parameters:
    - df: DataFrame with articles from API

    Returns:
    - df: Preprocessed DataFrame
    """

    df = df.copy()

    # Handle missing values
    df['text'] = df['text'].fillna('')
    df['title'] = df['title'].fillna('') if 'title' in df.columns else ''
    df['description'] = df['description'].fillna('') if 'description' in df.columns else ''
    df['content'] = df['content'].fillna('') if 'content' in df.columns else ''

    # Create combined text (if not already done)
    if 'combined_text' not in df.columns:
        # Combine title, description, and content
        text_parts = []
        if 'title' in df.columns:
            text_parts.append(df['title'].fillna(''))
        if 'description' in df.columns:
            text_parts.append(df['description'].fillna(''))
        if 'content' in df.columns:
            text_parts.append(df['content'].fillna(''))

        df['combined_text'] = text_parts[0]
        for part in text_parts[1:]:
            df['combined_text'] = df['combined_text'] + ' ' + part

    # Text features
    df['text_length'] = df['text'].str.len()
    df['word_count'] = df['text'].str.split().str.len()

    # Time features (if timestamp exists)
    if 'timestamp' in df.columns and df['timestamp'].notna().any():
        if not pd.api.types.is_datetime64_any_dtype(df['timestamp']):
            df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')

        df['hour'] = df['timestamp'].dt.hour
        df['day_of_week'] = df['timestamp'].dt.dayofweek
        df['month'] = df['timestamp'].dt.month

    return df

# ============================================
# 4. GET PROBABILITIES FROM MODEL
# ============================================

def get_probabilities(model, X):
    """
    Extract probabilities from MultiOutputClassifier

    Parameters:
    - model: Trained model
    - X: Feature matrix

    Returns:
    - probs: Probability matrix (samples x classes)
    """

    probs_list = []
    for estimator in model.estimators_:
        prob = estimator.predict_proba(X)[:, 1]
        probs_list.append(prob)

    probs = np.column_stack(probs_list)
    return probs

# ============================================
# 5. TOP-K PREDICTION
# ============================================

def predict_top_k(probs, k_min=1, k_max=3, min_prob=0.30):
    """
    Predict top K labels using probability ranking

    Parameters:
    - probs: Probability matrix
    - k_min: Minimum labels to predict
    - k_max: Maximum labels to predict
    - min_prob: Minimum probability threshold

    Returns:
    - preds: Binary prediction matrix
    """

    n_samples, n_labels = probs.shape
    preds = np.zeros_like(probs, dtype=int)

    for i in range(n_samples):
        sample_probs = probs[i]
        sorted_indices = np.argsort(sample_probs)[::-1]
        sorted_probs = sample_probs[sorted_indices]

        # Determine how many labels to predict
        selected_count = k_min

        for k in range(k_min, k_max):
            if k >= len(sorted_probs):
                break

            if sorted_probs[k] >= min_prob:
                if k == 0 or (sorted_probs[k-1] - sorted_probs[k]) <= 0.10:
                    selected_count = k + 1
                else:
                    break
            else:
                break

        selected_indices = sorted_indices[:selected_count]
        preds[i, selected_indices] = 1

    return preds

# ============================================
# 6. CLASSIFY ARTICLES
# ============================================

def classify_articles(df, model, mlb, vectorizer, k_min=1, k_max=3, min_prob=0.30):
    """
    Complete classification pipeline for new articles

    Parameters:
    - df: DataFrame with articles from API
    - model: Trained model
    - mlb: MultiLabelBinarizer
    - vectorizer: TfidfVectorizer
    - k_min, k_max, min_prob: Top-K parameters

    Returns:
    - df: DataFrame with predictions added
    """

    print("="*80)
    print("CLASSIFYING ARTICLES")
    print("="*80)

    if df.empty:
        print(" No articles to classify")
        return df

    # Step 1: Preprocess
    print(f"\n[1/4] Preprocessing {len(df)} articles...")
    df = preprocess_for_prediction(df)

    # Step 2: Transform text to features
    print("[2/4] Extracting features...")
    X = vectorizer.transform(df['combined_text'])

    # Step 3: Get probabilities
    print("[3/4] Computing probabilities...")
    probs = get_probabilities(model, X)

    # Step 4: Predict with Top-K
    print(f"[4/4] Making predictions (k_min={k_min}, k_max={k_max}, min_prob={min_prob})...")
    preds = predict_top_k(probs, k_min=k_min, k_max=k_max, min_prob=min_prob)
    pred_labels = mlb.inverse_transform(preds)

    # Add predictions to DataFrame
    df['predicted_categories'] = [list(labels) for labels in pred_labels]
    df['num_categories'] = [len(labels) for labels in pred_labels]

    # Add top probabilities for each article
    df['top_probabilities'] = [
        dict(zip(mlb.classes_[np.argsort(probs[i])[::-1][:5]],
                 sorted(probs[i])[::-1][:5]))
        for i in range(len(df))
    ]

    print("\n" + "="*80)
    print("CLASSIFICATION COMPLETE")
    print("="*80)
    print(f"Total articles classified: {len(df)}")
    print(f"Average categories per article: {df['num_categories'].mean():.2f}")
    print(f"\nCategory distribution:")
    all_categories = [cat for cats in df['predicted_categories'] for cat in cats]
    category_counts = pd.Series(all_categories).value_counts()
    print(category_counts)

    return df

# ============================================
# 7. DISPLAY RESULTS
# ============================================

def display_predictions(df, n_samples=10):

    print("\n" + "="*80)
    print("SAMPLE PREDICTIONS")
    print("="*80)

    for i in range(min(n_samples, len(df))):
        article = df.iloc[i]

        print(f"\n{'='*80}")
        print(f"Article {i+1}:")
        print(f"Title: {article.get('title', 'N/A')}")
        print(f"Source: {article.get('source_name', 'N/A')}")
        print(f"Published: {article.get('publishedAt', 'N/A')}")
        print(f"\nText preview: {article['text'][:150]}...")

        print(f"\n PREDICTED CATEGORIES: {article['predicted_categories']}")
        print(f"  Number of categories: {article['num_categories']}")

        print(f"\nTop-5 Probabilities:")
        for cat, prob in list(article['top_probabilities'].items())[:5]:
            marker = "" if cat in article['predicted_categories'] else " "
            print(f"  {marker} {cat:<20s}: {prob:.3f}")
        print("-"*80)

# ============================================
# 8. SAVE PREDICTIONS
# ============================================

def save_predictions(df, filepath='classified_articles.csv', include_probs=False):

    df_save = df.copy()

    df_save['predicted_categories'] = df_save['predicted_categories'].apply(lambda x: ','.join(x))

    if not include_probs:
        if 'top_probabilities' in df_save.columns:
            df_save = df_save.drop('top_probabilities', axis=1)
    else:
        df_save['top_probabilities'] = df_save['top_probabilities'].astype(str)

    try:
        df_save.to_csv(filepath, index=False)
        print(f" Predictions saved to {filepath}")
    except Exception as e:
        print(f" Error saving predictions: {e}")

# ============================================
# 9. MAIN CLASSIFICATION PIPELINE
# ============================================

def classify_from_api_data(df, model_path='news_classifier_model.pkl',
                           k_min=1, k_max=3, min_prob=0.30,
                           save_results=True, output_file='classified_articles.csv'):

    print("="*80)
    print("LOADING MODEL AND CLASSIFYING ARTICLES")
    print("="*80)

    model, mlb, vectorizer = load_model(model_path)

    if model is None:
        print(" Failed to load model. Aborting.")
        return df

    df = classify_articles(df, model, mlb, vectorizer, k_min, k_max, min_prob)

    display_predictions(df, n_samples=10)

    if save_results:
        print(f"\nSaving results to {output_file}...")
        save_predictions(df, output_file)

    return df

print("\nModel Classifier loaded!")
print("\nMain functions:")
print("  1. save_model(model, mlb, vectorizer) - Save trained model")
print("  2. load_model(filepath) - Load saved model")
print("  3. classify_from_api_data(df) - Complete classification pipeline")
```
