```python
# ============================================
# ğŸ”¥ GNews + News Classifier Integration Script
# ============================================

import requests
import pandas as pd
import joblib
import numpy as np
import scipy.sparse as sp

# --- 1ï¸âƒ£ Load Trained Model Bundle ---
bundle = joblib.load("/content/news_classifier_model.pkl")
model, mlb, vectorizer = bundle[0], bundle[1], bundle[2]
print("âœ… Model bundle loaded successfully!")

# --- 2ï¸âƒ£ Fetch News from GNews.io ---
API_KEY = "172285a1b5b6f981c517e59461b31a9a"   # ğŸ”‘ Replace with your API key
query = input("Enter a news topic (e.g., 'technology', 'sports', 'finance'): ").strip()
url = f"https://gnews.io/api/v4/search?q={query}&lang=en&country=us&max=10&apikey={API_KEY}"

response = requests.get(url)
data = response.json()

# --- 3ï¸âƒ£ Parse JSON Articles into DataFrame ---
articles = []
for item in data.get("articles", []):
    title = item.get("title", "")
    description = item.get("description", "")
    content = item.get("content", "")
    text = f"{title}. {description}. {content}".strip()
    if text:
        articles.append({
            "title": title,
            "description": description,
            "content": content,
            "text": text
        })

df_articles = pd.DataFrame(articles)
print(f"âœ… Retrieved {len(df_articles)} articles for query: {query}")

if df_articles.empty:
    raise ValueError("âš ï¸ No articles found. Try another query or check your API key.")

# --- 4ï¸âƒ£ Transform Text using Saved Vectorizer ---
X = vectorizer.transform(df_articles["text"])

# --- 5ï¸âƒ£ Auto-Fix Feature Mismatch (Padding/Trimming) ---
expected_features = model.n_features_in_
current_features = X.shape[1]

if current_features < expected_features:
    diff = expected_features - current_features
    print(f"âš™ï¸ Padding {diff} missing features...")
    X = sp.hstack([X, sp.csr_matrix((X.shape[0], diff))])

elif current_features > expected_features:
    diff = current_features - expected_features
    print(f"âš™ï¸ Trimming {diff} extra features...")
    X = X[:, :expected_features]

print(f"âœ… Feature alignment done: {X.shape[1]} features match model input.")

# --- 6ï¸âƒ£ Predict Labels ---
try:
    probs_list = [est.predict_proba(X)[:, 1] for est in model.estimators_]
    probs = np.column_stack(probs_list)
except AttributeError:
    probs = model.predict_proba(X)

threshold = 0.3
pred_labels = mlb.inverse_transform(probs > threshold)

# --- 7ï¸âƒ£ Append Predictions ---
df_articles["predicted_labels"] = pred_labels
df_articles["top_k_labels"] = [
    [mlb.classes_[i] for i in np.argsort(p)[::-1][:3]]
    for p in probs
]

# --- 8ï¸âƒ£ Display Results ---
for _, row in df_articles.iterrows():
    print("\nâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
    print(f"ğŸ“° Title: {row['title']}")
    print(f"ğŸ’¬ Description: {row['description'][:100]}...")
    print(f"ğŸ”– Predicted Labels: {row['predicted_labels']}")
    print(f"â­ Top 3 Labels: {row['top_k_labels']}")

# --- 9ï¸âƒ£ (Optional) Save Results ---
df_articles.to_csv("classified_news_results.csv", index=False)
print("\nğŸ“ Results saved to classified_news_results.csv")
```

```python
# ==============================================================
# ğŸ“° GNews + News Classifier + Save to CSV (Complete Script)
# ==============================================================

import requests
import pandas as pd
import joblib
import numpy as np
import scipy.sparse as sp

# --- 1ï¸âƒ£ Load Trained Model ---
bundle = joblib.load("/content/news_classifier_model.pkl")
model, mlb, vectorizer = bundle[0], bundle[1], bundle[2]
print("âœ… Model bundle loaded successfully!")

# --- 2ï¸âƒ£ Fetch News from GNews.io ---
API_KEY = "172285a1b5b6f981c517e59461b31a9a"   # ğŸ”‘ Replace with your API key
query = input("Enter a news topic (e.g., 'technology', 'sports', 'finance'): ").strip()

url = f"https://gnews.io/api/v4/search?q={query}&lang=en&country=us&max=10&apikey={API_KEY}"
response = requests.get(url)
data = response.json()

# --- 3ï¸âƒ£ Parse JSON to DataFrame (Keep All Fields) ---
articles = []
for item in data.get("articles", []):
    source = item.get("source", {})
    articles.append({
        "title": item.get("title", ""),
        "description": item.get("description", ""),
        "content": item.get("content", ""),
        "url": item.get("url", ""),
        "image": item.get("image", ""),
        "publishedAt": item.get("publishedAt", ""),
        "source_name": source.get("name", ""),
        "source_url": source.get("url", ""),
        "language": item.get("language", "en"),  # default
        "country": item.get("country", "us"),    # default
    })

df_articles = pd.DataFrame(articles)
print(f"âœ… Retrieved {len(df_articles)} articles for query: {query}")

if df_articles.empty:
    raise ValueError("âš ï¸ No articles found. Try another query or check your API key.")

# --- 4ï¸âƒ£ Prepare Text for Vectorizer ---
df_articles["text"] = (
    df_articles["title"].fillna("") + ". " +
    df_articles["description"].fillna("") + ". " +
    df_articles["content"].fillna("")
)

# --- 5ï¸âƒ£ Vectorize Text ---
X = vectorizer.transform(df_articles["text"])

# --- 6ï¸âƒ£ Fix Feature Size Mismatch ---
expected_features = model.n_features_in_
current_features = X.shape[1]

if current_features < expected_features:
    diff = expected_features - current_features
    print(f"âš™ï¸ Padding {diff} missing features...")
    X = sp.hstack([X, sp.csr_matrix((X.shape[0], diff))])

elif current_features > expected_features:
    diff = current_features - expected_features
    print(f"âš™ï¸ Trimming {diff} extra features...")
    X = X[:, :expected_features]

print(f"âœ… Feature alignment done: {X.shape[1]} features match model input.")

# --- 7ï¸âƒ£ Predict Labels ---
try:
    probs_list = [est.predict_proba(X)[:, 1] for est in model.estimators_]
    probs = np.column_stack(probs_list)
except AttributeError:
    probs = model.predict_proba(X)

threshold = 0.3
pred_labels = mlb.inverse_transform(probs > threshold)

# --- 8ï¸âƒ£ Append Predictions ---
df_articles["predicted_labels"] = [", ".join(labels) for labels in pred_labels]
df_articles["top_3_labels"] = [
    ", ".join([mlb.classes_[i] for i in np.argsort(p)[::-1][:3]]) for p in probs
]

# --- 9ï¸âƒ£ Save Everything to CSV ---
output_path = f"/content/{query}_classified_news.csv"
df_articles.to_csv(output_path, index=False, encoding="utf-8-sig")
print(f"\nğŸ“ All results saved successfully to: {output_path}")

# --- ğŸ”Ÿ Preview Results ---
print("\nğŸ§¾ Sample Output:")
display(df_articles[["title", "predicted_labels", "top_3_labels"]].head())
```
